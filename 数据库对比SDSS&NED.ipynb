本代码跳过第二段（在服务器上运行），是将下载后的、与SLOAN及NED数据对比过的HERSCHEL数据中的‘STAR’类型重新与NED中记录的相应类型做对比，进行初步的修正。因为有些SLOAN中的‘STAR’实际上是其他类型的天体。

```python
import numpy as np
import pandas as pd
import astropy.units as u   #单位
from astropy.coordinates import SkyCoord  #天球坐标
from astroquery.sdss import SDSS   #Sloan数值巡天
from astroquery.ned import Ned  #NASA/IPAC Extragalactic Database
```

```python
data_original=pd.read_csv('irsa_catalog_search_results.tbl',sep='\s+',skiprows=32,header=None,names=['name','band','ra','dec','raerr','decerr','flux','type'])
#data_original=pd.read_csv('text.tbl',sep='\s+',header=None,names=['name','band','ra','dec','raerr','decerr','flux','type'])
index=data_original.index.values  #序号

#<MaskedColumn name='Type' dtype='object' description="NED's Preferred Object Type: G,GPair,GTrpl,GGroup,GClstr,QSO,AbLS,RadioS,IrS,EmLS,UvES,XrayS,SN"
for i in index:
    print(i)
    ra=data_original['ra'][i]
    dec=data_original['dec'][i]
    try:
        table_sdss=SDSS.query_crossid(SkyCoord(ra,dec,unit=(u.deg,u.deg)),radius=3*u.arcsec,data_release=15)    #第15次释放数据
        if table_sdss==None:
            table_ned=Ned.query_region(SkyCoord(ra,dec,unit=(u.deg,u.deg)),radius=3*u.arcsec,equinox='J2000.0')   #2000年坐标
            if len(table_ned)>0:
            #data_original['type'][i]=table_ned['Type'].description 
                data_original['type'][i]=table_ned['Type'][0]   
            else:
                continue
        else:
            data_original['type'][i]=table_sdss['type'][0]
    except:
        continue
data_original.to_csv('text.tbl',float_format="%.6f",index=0)
# 输出text.tbl
```

```python
#修正text
data_test = pd.read_csv("text.tbl")
type_test = data_test['type']
ra_test   = data_test['ra']
dec_test  = data_test['dec']
num = len(type_test)


a = 0
for i in range(num):
    try:
        if type_test[i] == "STAR":
            a+=1
            table_ned=Ned.query_region(SkyCoord(ra_test[i],dec_test[i],unit=(u.deg,u.deg)),radius=3*u.arcsec,equinox='J2000.0')
            if len(table_ned)>0:
                data_test['type'][i]=table_ned['Type'][0]
            else:
                continue
    except:
        continue
        
print(a)
data_test.to_csv('text_test.csv',float_format="%.6f",index=0)
```
