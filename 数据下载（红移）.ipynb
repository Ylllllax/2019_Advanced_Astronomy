```python
import numpy as np
import pandas as pd
import warnings
import time
import astropy.units as u   #单位
from astropy.coordinates import SkyCoord   # 天球坐标
from astroquery.irsa import Irsa  # 所需数据库 
from astroquery.ned import Ned  #NASA/IPAC Extragalactic Database
```

```python
data = pd.read_csv("flux_original.csv")
data['redshift'] = np.nan
```

```python
# 寻找所需红移数据

index = len(data)

for i in range(0, index):  # 修改range初值，可从上次中断保存处继续运行
    print ('The echo is', i, '/', index)
    ra = data['ra'][i]
    dec = data['dec'][i]
    warnings.filterwarnings('ignore')
    
    
    table_ned = Ned.query_region(SkyCoord(ra, dec, unit=(u.deg, u.deg)), 
                             radius=3*u.arcsec, equinox='J2000.0')
    if len(table_ned) > 0:
        data['redshift'][i] = table_ned['Redshift'][0]

    '''
    # 定时中断保存
    interval = 200
    if (i % interval) == 0 and i != 0 :   # 每查询200次保存一次文件
        data_name = 'flux_redshift_original_%s.csv' % (i)
        data.to_csv(data_name, float_format="%.6f", index=0)
        print ('The file %s has been saved.\n' % data_name)
        time.sleep(5)   # 提示信息延时5秒
    '''
data.to_csv('flux_redshift_original.csv', float_format="%.6f", index=0)
```
