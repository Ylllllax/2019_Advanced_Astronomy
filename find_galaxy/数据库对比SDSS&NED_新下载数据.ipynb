本代码是将下载后的、与SLOAN及NED数据对比过的HERSCHEL数据中的‘STAR’类型重新与NED中记录的相应类型做对比，进行初步的修正。因为有些SLOAN中的‘STAR’实际上是其他类型的天体。

```python
import numpy as np
import pandas as pd
import astropy.units as u   #单位
from astropy.coordinates import SkyCoord  #天球坐标
from astroquery.sdss import SDSS   #Sloan数值巡天
from astroquery.ned import Ned  #NASA/IPAC Extragalactic Database
```

```python
#修正text
data_test2 = pd.read_csv("text2.tbl")
type_test2 = data_test2['type']
ra_test2   = data_test2['ra']
dec_test2  = data_test2['dec']
num = len(type_test2)


a = 0
for i in range(num):
    try:
        if type_test2[i] == "STAR":
            a+=1
            table_ned=Ned.query_region(SkyCoord(ra_test2[i],dec_test2[i],unit=(u.deg,u.deg)),radius=3*u.arcsec,equinox='J2000.0')
            if len(table_ned)>0:
                data_test2['type'][i]=table_ned['Type'][0]
            else:
                continue
    except:
        continue
        
print(a)
data_test2.to_csv('text_test2.csv',float_format="%.6f",index=0)
```

```python
#验证
data_ensure = pd.read_csv("text_test2.csv")
type_ensure = data_ensure['type']
ra_ensure   = data_ensure['ra']
dec_ensure  = data_ensure['dec']

b = 0
for i in range(num):
    try:
        if type_ensure[i] == "STAR":
            b+=1
    except:
        continue
        
print(b)
```
